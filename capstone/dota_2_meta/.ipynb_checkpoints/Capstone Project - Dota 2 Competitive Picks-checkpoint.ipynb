# Dota 2 Competitive Picks

## Introduction

### Problem Statement
Periodically, game developers release major patches that alter the way the game is played in order to keep the game interesting. As casual gamers, we may not possess the knowledge to comprehend how these patches affect the game playstyle and its meta (the most effective tactics available) heroes. As a result, picking the wrong hero might cost us a game.

### Objective
Therefore, by examining the hero picks within the professional gaming scene, we can gain insights into the meta and which powerful heroes provide us with the best opportunity to secure victories in our games.

## Data Gathering and Preparation

```python
# Import  relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data and data cleaning
pro_df = pd.read_csv('./datasets/Current_Pro_meta.csv')
pro_df.head()
```

```python
# Check for any null values
print(pro_df.info())
pro_df.isnull().sum()
```

```python
# Remove Unnamed: 0 column
pro_df.drop(columns=['Unnamed: 0'], axis = 1, inplace = True)
pro_df.head()
```

```python
# Convert boolean column "Niche Hero?" to numeric
pro_df["Niche Hero?"] = pro_df["Niche Hero?"].astype(int)
pro_df['Niche Hero?']
```

```python
# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
pro_df.head()
```

## Explotary Data Analysis (EDA)

### 1. Times Picked, Times Banned and Win Rate analysis

```python
plt.style.use('fast')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,6))

ax[0,0].hist(pro_df['Times Picked'], color = 'c')
ax[0,0].set_title('Times Picked Distribution')
ax[0,0].set_xlabel('Times Picked')
ax[0,0].set_ylabel('count')


ax[0,1].hist(pro_df['Times Banned'] , color = 'm')
ax[0,1].set_title('Times Banned Distribution')
ax[0,1].set_xlabel('Times Banned')
ax[0,1].set_ylabel('count')

ax[1,0].scatter(pro_df['Times Picked'], pro_df['Win Rate'], color = 'c')
ax[1,0].set_title('Win Rate vs Times Picked')
ax[1,0].set_xlabel('Times Picked')
ax[1,0].set_ylabel('Win Rate')
ax[1,0].set_xlim(0)

ax[1,1].scatter(pro_df['Times Banned'], pro_df['Win Rate'], color = 'm' )
ax[1,1].set_title('Win Rate vs Times Banned')
ax[1,1].set_xlabel('Times Banned')
ax[1,1].set_ylabel('Win Rate')
ax[1,1].set_xlim(0)

fig.tight_layout()
plt.show()
```

- Majority of heroes are picked or banned less than 400.
- There seems to be a weak correlation between Win Rate, and the Time Picked and Times Banned.

### 2. Roles Analysis

```python
# Analyze the distribution of heroes across different roles
roles = pro_df["Roles"].str.split(", ", expand=True).stack()
role_counts = roles.value_counts()
role_counts
```

```python
# Roles popularity
Roles = {'Carry':0, 'Support':0, 'Disabler':0, 'Nuker':0, 'Escape':0, 'Durable':'', 'Pusher':0, 'Initiator':0}
for x in Roles.keys():
    role_df = pro_df[pro_df['Roles'].str.contains(x)]
    count = role_df['Times Picked'].mean()
    Roles[x] = count
sortedRoles = sorted(Roles)
print(Roles)
```

```python
# Sort Roles dict by values
import operator
sorted_Roles = dict( sorted(Roles.items(), key=operator.itemgetter(1),reverse=True)) # descending order

keys = list(sorted_Roles.keys())
values = list(sorted_Roles.values())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

sns.barplot(ax=ax[0], x=role_counts.values, y=role_counts.index, orient='h', 
            palette=['magenta' if (x=='Support')|(x=='Carry') else 'c' for x in role_counts.index])
ax[0].set_title('Heroes Quantity for Each Role')
ax[0].set_xlabel('Heroes Quantity')
ax[0].set_ylabel('Roles')


sns.barplot(ax=ax[1], x=values, y=keys, orient='h', 
            palette=['magenta' if (x=='Support')|(x=='Carry') else 'c' for x in keys])
ax[1].set_title('Roles Popularity')
ax[1].set_xlabel('Times Picked Average')
ax[1].set_ylabel('Roles')

fig.tight_layout()
plt.show()
```

- There seems like a preference in hero design with disablers being designed the most and pushers being designed the least.
- Disablers being most popular. Probably because highest quantity of heroes.
- Support role are picked disproportionately often, despite the low amount of heroes.
- Carry role are picked disproportionately less than Support, despite having higher heroes quantity.

## KMeans Clustering

The goal is to identify the most contested heroes and understand what are the traits of these heroes. Therefore, clustering is chosen.

```python
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
```

```python
# New df for Clustering by removing 'Name' and 'Roles' column
num_df = pro_df.copy()
num_df.drop(['Name','Roles'], axis=1,inplace=True)
num_df.head()
```

```python
# Label encoding for column 'Primary Attribute' and 'Attack Type'
le = LabelEncoder()
num_df['Primary Attribute']=le.fit_transform(num_df['Primary Attribute'])
num_df['Attack Type']=le.fit_transform(num_df['Attack Type'])
num_df.head()
```

```python
# Determine number of cluster

inertias = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(num_df)
    inertias.append(kmeans.inertia_) #main function to determine the clusters i "inertia"

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Optimal K=4
# Fit K-Means to the data with K=4
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(num_df)

# Add the cluster labels to the DataFrame
num_df['Cluster'] = clusters

# Create separate dataframes for each cluster
cluster_dataframes = [num_df[num_df['Cluster'] == i] for i in range(4)]
```

```python
num_df.head()
```

```python
# Visualize the clusters
for i, cluster_df in enumerate(cluster_dataframes):
    plt.scatter(cluster_df['Times Picked'], cluster_df['Times Banned'], label=f'Cluster {i + 1}')

plt.xlabel('Times Picked')
plt.ylabel('Times Banned')
plt.title('K-Means clustering (K=4)')
plt.legend()
plt.show()

for i, cluster_df in enumerate(cluster_dataframes):
    plt.scatter(cluster_df['Times Picked'], cluster_df['Attack Range'], label=f'Cluster {i + 1}')

plt.xlabel('Times Picked')
plt.ylabel('Attack Range')
plt.title('K-Means clustering (K=4)')
plt.legend()
plt.show()
```

- Cluster 1 and 4 are most contested due to high pick and ban rate.
- Difference between 1 and 4 is attack range, 4 are ranged heroes while 1 are melee or short ranged heroes

```python
# Merge dataframe

col_to_add = num_df.columns.difference(pro_df.columns)
merged = pd.merge(pro_df, num_df[col_to_add], how='outer', left_index=True, right_index=True)
```

```python
def cluster(cluster):
    return cluster + 1
merged['Cluster'] = merged['Cluster'].apply(cluster)
```

### Cluster 1 and 4 are the most contested heroes in pro scene

```python
meta_df = merged[(merged['Cluster']==1)|(merged['Cluster']==4)]
meta_df.reset_index(drop=True, inplace=True)
meta_df
```

```python
# Roles popularity
Roles = {'Carry':0, 'Support':0, 'Disabler':0, 'Nuker':0, 'Escape':0, 'Durable':0, 'Pusher':0, 'Initiator':0}
for x in Roles.keys():
    role_df = meta_df[meta_df['Roles'].str.contains(x)]
    count = role_df['Times Picked'].mean()
    Roles[x] = count
sortedRoles = sorted(Roles)
print(Roles)
```

```python
sorted_Roles = dict( sorted(Roles.items(), key=operator.itemgetter(1),reverse=True)) # descending order

keys = list(sorted_Roles.keys())
values = list(sorted_Roles.values())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

sns.barplot(ax=ax[0], x=role_counts.values, y=role_counts.index, orient='h', 
            palette=['magenta' if (x=='Support')|(x=='Durable') else 'c' for x in role_counts.index])
ax[0].set_title('Heroes Quantity for Each Role')
ax[0].set_xlabel('Heroes Quantity')
ax[0].set_ylabel('Roles')

sns.barplot(ax=ax[1], x=values, y=keys, orient='h', 
            palette=['magenta' if (x=='Support')|(x=='Durable') else 'c' for x in keys])
ax[1].set_title('Roles Popularity')
ax[1].set_xlabel('Average Times Picked')
ax[1].set_ylabel('Roles')

fig.tight_layout()
plt.show()
```

## Recommendations

```python
top_wr = meta_df[meta_df['Win Rate']>50]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))

sns.barplot(x=top_wr['Win Rate'], y=top_wr['Name'], orient='h',
            order=top_wr.sort_values('Win Rate',ascending=False).Name,
           palette='crest')
ax.set_title('Top Heroes')
ax.set_xlabel('Win Rate')
ax.set_ylabel('Name')
plt.show()
```

- If you want to win more games, these are the most contested heroes by pro players with Win Rate more than 50%.
